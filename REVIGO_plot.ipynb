```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
matplotlib.style.use("ggplot")

df_revigo = pd.read_csv('./REVIGO_plot.csv')

m = (df_revigo['plot_X'] != 'null') & (df_revigo['plot_Y'] != 'null')
df_revigo_nn = df_revigo[m]

m = (df_revigo_nn['dispensability'] < 0.05)
df_revigo_nn_repr = df_revigo_nn[m]

color = df_revigo_nn['plot_Y'].values.astype(float)

fig,ax = plt.subplots(1,figsize=(8,8))
ax.scatter(df_revigo_nn['plot_X'],df_revigo_nn['plot_Y'],s=df_revigo_nn['plot_size']*10,c=color)
for i in df_revigo_nn_repr.index:
    edf = df_revigo_nn_repr.loc[i]
    x = edf['plot_X']
    y = edf['plot_Y']
    text = edf['description']
    ax.text(x,y,text)
ax.set_xlabel('semantic X')
ax.set_ylabel('semantic Y')
plt.savefig('revigo.png',dpi=400)
```
